# Applying Filters:

```python
#Filtering using number (showing all the entries with rating more than 4):
data[data['rating']>4]
```

```python
#Filtering on the basis of colour (using same logic as above):
color=['Silver']
data[data['color'].isin(color)]
```

```python
#Filtering on the basis of storage:
storage=['256 GB']
data[data['harddisk'].isin(storage)]
```
