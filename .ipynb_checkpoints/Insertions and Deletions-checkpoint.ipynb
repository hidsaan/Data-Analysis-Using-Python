```python
students={"Sno.":['1', '2', '3', '4', '5'],
          "Name":['Sehreen', 'Hidsaan', 'Zarnigah', 'Falah', 'Nargis'],
          "Roll no.":['1', '2', '3', '4', '5'],
          "Subjects":['Asp.net', 'Python', 'React', 'c++', 'java']}

import pandas as pd

data= pd.DataFrame(students)
data
```

```python
#insertion of a new column with values:
'''
data.insert(4,'Address',['lal chowk','nishat','lal bazar','hawal','hmt'])'''
data
```

```python
#inserting a new column with duplicate values from an already present column:
data.insert(3,'Courses',data['Subjects'])
data
```

```python
#deletion using pop:
data.pop('Courses')
data
```

```python
#deletion using del:
del data['Address']
data
```

```python
#exporting in csv form:
data.to_csv('students.csv')
```

```python
#exporting in xlsx form
data.to_excel('studentstable.xlsx')
```

```python
#Turning the table into html webpage:
mytable=data.to_html()
mytable
```
